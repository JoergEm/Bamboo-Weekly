```r
# Imports
if (!require("readxl")) install.packages("readxl")
if (!require("httr")) install.packages("httr")
if (!require("dplyr")) install.packages("dplyr")

library(readxl)
library(httr)
library(dplyr)

cat("Imports âœ…\n")
```

```r
# Function creating local folders
create_folders <- function(folders) {
  for (folder in folders) {
    if (!dir.exists(folder)) {
      dir.create(folder)
    }
  }
  cat("Folders âœ…\n")
}
```

```r
# Function downloading data locally
download_data <- function(url, filepath) {
  if (file.exists(filepath)) {
    cat("Data loaded from existing file ðŸ“\n")
    return()
  }

  tryCatch({
    response <- httr::GET(url)
    writeBin(httr::content(response, "raw"), filepath)
    cat("Data âœ…\n")
  }, error = function(e) {
    cat("Error âŒ:", e$message, "\n")
  })
}
```

```r
# Links and folders to recieve data and read into DataFrame
url <- "https://images.transparencycdn.org/images/CPI2022_GlobalResultsTrends.xlsx"
filepath <- file.path("data", "CPI2022_GlobalResultsTrends.xlsx")
folders <- c("data", "results")

create_folders(folders)
download_data(url, filepath)

data <- read_excel(filepath, sheet = "CPI 2022 (final)", skip = 2)  # skip = 2 Ã¼berspringt die ersten 2 Zeilen
data <- as.data.frame(data)
```
