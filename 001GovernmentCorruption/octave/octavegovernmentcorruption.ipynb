```octave
% Function creating local folders
function success = create_folders(folders)
  success = false;
  try
    for k = 1:numel(folders)
      folderpath = fullfile(pwd, folders{k});
      if ~exist(folderpath, 'dir')
        mkdir(folderpath);
      end
    end
    fprintf("Folders ✅\n");
    success = true;
  catch
    fprintf("Error ❌\n");
    success = false;
  end
end
```

```octave
% Function downloading data locally
function success = download_data(url, filename)
  success = false;
  try
    urlwrite(url, filename);
    success = true;
  catch
    fprintf("Error ❌\n", url);
    success = false;
  end
end
```

```octave
% Links and folders to recieve data and read into DataFrame
url = 'https://images.transparencycdn.org/images/CPI2022_GlobalResultsTrends.xlsx';
filename = 'CPI2022_GlobalResultsTrends.xlsx';
folders = {"data", "results"};
filepath = fullfile(folders{1}, filename);
create_folders(folders);

if ~exist(filepath, 'file')
    if download_data(url, filepath)
    else
        fprintf("Error ❌\n");
    end
else
    fprintf("Data loaded from existing file ✅\n");
end

if exist(filepath, 'file')
    try
        data = xlsread(filepath, 'CPI 2022 (final)');
        if size(data,1) > 2
            data(1:2, :) = [];
        end
        fprintf("Data ✅\n");
    catch
        fprintf("Error ❌\n");
    end
end
```

```octave
% According to Transparency International, what five countries were least corrupt in 2022?
try
    [num, txt, raw] = xlsread(filepath, 'CPI 2022 (final)');

    % Skiprows
    raw(1:2,:) = [];
    rank_col = find(strcmp(raw(1,:), 'Rank'));
    country_col = find(strcmp(raw(1,:), 'Country / Territory'));
    ranks = cell2mat(raw(2:end, rank_col));
    countries = raw(2:end, country_col);
    [~, sort_idx] = sort(ranks);
    top5_countries = countries(sort_idx(1:5));
    fprintf("Die 5 am wenigsten korrupten Länder 2022:\n");
    disp(top5_countries); 
catch
    fprintf("Error ❌\n");
end
```

```octave
% According to the same data, what five countries were most corrupt in 2022?
try
    [num, txt, raw] = xlsread(filepath, 'CPI 2022 (final)');
    
    % Skiprows
    raw(1:2,:) = [];
    rank_col = find(strcmp(raw(1,:), 'Rank'));
    country_col = find(strcmp(raw(1,:), 'Country / Territory'));
    ranks = cell2mat(raw(2:end, rank_col));
    countries = raw(2:end, country_col);
    [~, sort_idx] = sort(ranks);
    tail5_countries = countries(sort_idx(end-4:end));
    disp(tail5_countries);
catch
    fprintf("Error ❌\n");
end
```
