```julia
# Imports
using Pkg
Pkg.add("XLSX")
Pkg.add("HTTP")
Pkg.add("DataFrames")
using HTTP, XLSX, DataFrames
using Base.Filesystem: isfile
println("Imports ‚úÖ");
```

```julia
# Function creating local folders
function create_folders(folders::Vector{String})
    for folder in folders
        if !isdir(folder)
            mkdir(folder)
        end
    end
    println("Folders ‚úÖ")
end;
```

```julia
# Function downloading data locally
function download_data(url, filepath)
    if isfile(filepath)
        println("Data loaded from existing file üìÅ")
        return
    end
    try
        response = HTTP.get(url)
        open(filepath, "w") do f
            write(f, response.body)
        end
        println("Data ‚úÖ")
    catch error
        println("Error ‚ùå: $error")
    end
end;
```

```julia
#  Links and folders to recieve data and read into DataFrame
url = "https://images.transparencycdn.org/images/CPI2022_GlobalResultsTrends.xlsx"
filepath = joinpath("data", "CPI2022_GlobalResultsTrends.xlsx")
folders= ["data", "results"]
create_folders(folders)
download_data(url, filepath)
data = DataFrame(XLSX.readtable(filepath,"CPI 2022 (final)",header=true,first_row=3))
```

```julia
# According to Transparency International, what five countries were least corrupt in 2022?
sort!(data, :Rank)
head = data[1:5, :]
```
